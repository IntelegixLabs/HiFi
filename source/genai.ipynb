```python
from dotenv import load_dotenv
load_dotenv()
import os
import google.generativeai as genai
import PIL.Image
from IPython.display import Image
```

```python
genai.configure(api_key = os.getenv("GEMINI_API_KEY"))
```

```python
img = PIL.Image.open('graph.jpeg')
display(img)

model = genai.GenerativeModel(model_name='gemini-pro-vision')
prompt ="""
        Describe what the graph in the image represents.
        """
response = model.generate_content([prompt, img])
print(response.text)
```

 The graph in the image represents the historical prices of a stock over time. The x-axis shows the time, and the y-axis shows the price. The green bars represent periods when the price increased, and the red bars represent periods when the price decreased. The size of the bar represents the magnitude of the price change. The graph shows that the stock price has been trending upward over the long term, but there have been some periods of volatility. The stock price reached a high in 2021 and then declined in 2022.

```python
prompt ="""
        Describe what the graph in the image represents. Briefly explain what the x and y-axis values ranges, 
        overall trend of the stock, fluctuations if any. 
        Based on the candlesticks predict the highest, lowest and most recent data points along with percentage rise and fall. 
        Suggest some future investment recommendations based on above data points.
        Do not try to make up answers
        """
response_vision = model.generate_content([prompt, img])
print(response_vision.text)
```

 The graph in the image is a candlestick chart of the stock price of a company over a 15-year period. The x-axis shows the time, and the y-axis shows the price. The green candlesticks indicate that the price increased over that period, and the red candlesticks indicate that the price decreased. The highest data point is at the top of the chart, and the lowest data point is at the bottom. The most recent data point is at the right of the chart.

The overall trend of the stock is upward, but there have been some fluctuations along the way. The stock price reached a high in 2014, and then it fell sharply in 2015. The stock price has been steadily rising since then.

Based on the candlesticks, the highest data point is 240, the lowest data point is 60, and the most recent data point is 186.34. The percentage rise from the lowest to highest data point is 300% and fall from highest to lowest data point is 75%.

Some future investment recommendations based on the above data points are:
- The stock price is currently trading at 186.34, which is below its 52-week high of 240. This could be a good opportunity to buy the stock at a discount.
- The stock has been steadily rising since 2015. This indicates that the company is doing well and that the stock price is likely to continue to rise.
- The stock price is volatile, so investors should be prepared for fluctuations in the price.
